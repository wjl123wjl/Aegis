```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  

import os
import numpy as np
import seaborn as sns
```

```python
Res_acc=np.array([[91.84,91.29,90.02,85.58],[92.21,90.98,90.44,84.88],[91.33,90.77,89.55,84.65],[88.36,np.nan,np.nan,np.nan]])

Res_mean=np.array([[28,46.4,58.79,43.5],[29.3,46.4,74.8,45.5],[40.8,55.3,70.4,45.3],[541.2,541.2,541.2,541.2]])

Res_std=np.array([[4.47,10.57,4.14,8.5],[2.38,11.43,15.59,8.734],[6.3,6.9,12.82,6.5],[49.8,49.8,49.8,49.8]])
```

```python
sns.set(style="whitegrid")
sns.set(style="ticks")
# sns.despine()
sns.set_style({"font.sans-serif":"DejaVu Sans"})
sns.set_style({"grid.color":'0.9'})
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 1})
```

```python
# ax = plt.axes()
df = pd.DataFrame(Res_acc)
mask = df.isnull()

ax = sns.heatmap(df, annot=True, square=True, fmt='.4g', mask=mask)
ax.set_title('ResNet20, Top-1 test accuracy of Clean model')

plt.ylabel('$N_q$', fontsize = 15)
plt.yticks(np.arange(0, 4)+0.5, (8,6,4,1))
plt.xlabel('$\lambda$', fontsize = 15)
plt.xticks(np.arange(0, 4)+0.5, (0,5e-4,1e-3,5e-3))

figure = ax.get_figure()    
figure.savefig('resnet_20_acc_nq_lambda.pdf', bbox_inches="tight", transparent=True)
```
